```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set()
df = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv", 
                 sep=";")
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='viridis')
```

```python
# beacuse the figure is complete purple it means there is no missing valye in the dataset
# If there was any missing value, the figure would have had different color shade on purple color background
```

```python
# TO CHECK CORRELATION
```

```python
plt.figure(figsize=(8,4))
sns.heatmap(df.corr(),cmap='Greens', annot = False)

```

```python

# cmap='Greens' sets the colormap of the heatmap to the Greens colormap.
#       This colormap is a sequential colormap that ranges from dark green to light green.
# annot=False disables annotations on the heatmap. 
#       Annotations are small text labels that are displayed on each cell of the heatmap.
```

```python
# trying to implment the same heatmap with annot = true
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(),cmap='Greens', annot = True)

#Changing the value of annot=True
#the output will show values by which features are correlated to each other in grid-cells.
```

```python
# in the above figure positive correlation is represented by lighter shades
```

```python
k = 12
cols = df.corr().nlargest(k, 'quality')['quality'].index
cm = df[cols].corr()
plt.figure(figsize=(12,8))
sns.heatmap(cm, annot=True, cmap = 'viridis')
```

```python
# From above we can see, there is a strong positive correlation of density with residual sugar.
# However, a strong negative correlation of density and alcohol.

# Also, there is no correlation between free sulphur dioxide and quality.
```
